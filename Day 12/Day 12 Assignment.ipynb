```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt

df = pd.read_csv("E:/LetsUpgrade AI-ML/Day 11/general_data.csv")
df.columns
```

```python
df.isnull()
df.dropna()
```

```python
df.drop_duplicates()
```

```python
df["Attrition"].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
df
```

```python
cat = [i for i in df.columns if df.dtypes[i] == 'object']
cat
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

df.Attrition = enc.fit_transform(df.Attrition)

df['BusinessTravel'] = enc.fit_transform(df['BusinessTravel'])
df['Department'] = enc.fit_transform(df['Department'])
df['EducationField'] = enc.fit_transform(df['EducationField'])
df['Gender'] = enc.fit_transform(df['Gender'])
df['JobRole'] = enc.fit_transform(df['JobRole'])
df['MaritalStatus'] = enc.fit_transform(df['MaritalStatus'])
df
```

```python
df_yes = df[df['Attrition']== 1]
df_no = df[df['Attrition']== 0]
```

## ManWhitney Test

```python
from scipy.stats import mannwhitneyu
stats, p = mannwhitneyu(df_yes.Age, df_no.Age)
print(stats,p)
```

here,
p = 2.9951588479067175e-30

p value is less than 0.05, so reject the null value and select the alternate value

```python
from scipy.stats import mannwhitneyu
stats, p = mannwhitneyu(df_yes.MonthlyIncome, df_no.MonthlyIncome)
print(stats,p)
```

here,
p = 0.053577283839938566

p value is greater than 0.05, so accept the null value.

## Chi Square test

```python
from scipy.stats import chi2_contingency
chitable = pd.crosstab(df.Attrition,df.MaritalStatus)
chitable
```

```python
stats,p,dof, expected = chi2_contingency(chitable)
print(stats,p)
```

here,
p = 8.45385940605786e-31

p value is less than 0.05, so reject the null value and select the alternate value

```python
chitable = pd.crosstab(df.Gender,df.Attrition)
chitable
```

```python
stats,p,dof, expected = chi2_contingency(chitable)
print(stats,p)
```

here,
p = 0.24529482862926827

p value is greater than 0.05, so accept the null value.
